```python
import random
import pygame #conda install -c cogsic pygame + install pygame with sudo apt-get install python-pygame
from  pygame.locals import *

pygame.init()

if not pygame.font:
    print("No fonts!")
   
myfont = pygame.font.SysFont("Arial", 16)
```

```python
screen_width = 500
screen_height = 500

win = pygame.display.set_mode((screen_width,screen_height))
pygame.display.set_caption("First Game")
```

```python
#coords (0,0) coin en haut à gauche
x = screen_width / 2
y = screen_height / 2
w = 10
h = 15
w_save = 10
h_save = 15
v = 1
```

```python
class Elipse:
    def __init__(self, fminx, fmaxx, fminy, fmaxy, width, height, velocity=-1, color=(255,0,0), x = 0, y = 0):
        self.minx = fminx
        self.maxx = fmaxx
        self.miny = fminy
        self.maxy = fmaxy
        self.w = width
        self.h = height
        
        if self.w < self.h:
            self.w_save = self.w
            self.h_save = self.h
        else:
            self.w_save = self.h
            self.h_save = self.w
        
        if velocity < 0:
            self.v = max(self.w, self.h)
        else:
            self.v = velocity
            
        self.color = color
        
        self.x = x
        self.y = y
        self.d = 1
        
    def getColor(self):
        return self.color
    
    def getCoords(self):
        return (self.x, self.y, self.w, self.h)
    
    def getCenter(self):
        return (self.x + (self.w / 2), self.y + (self.h / 2))
    
    def getHead(self):
        if self.d == 1:
            return (self.x + (self.w / 2), self.y)
        if self.d == 2:
            return (self.x + (self.w / 2), self.y + self.h)
        if self.d == 3:
            return (self.x, self.y + (self.h / 2))
        if self.d == 4:
            return (self.x + self.w, self.y + (self.h / 2))
    
    def up(self):
        if self.y > self.miny + self.h:
            self.y -= self.v
            if self.w != self.w_save:
                self.w = self.w_save
            if self.h != self.h_save:
                self.h = self.h_save
        self.d = 1
                
    def down(self):
        if self.y < self.maxy - self.h:
            self.y += self.v
            if self.w != self.w_save:
                self.w = self.w_save
            if self.h != self.h_save:
                self.h = self.h_save
        self.d = 2
                
    def left(self):
        if self.x > self.minx + self.w:
            self.x -= self.v
            if self.w != self.h_save:
                self.w = self.h_save
            if self.h != self.w_save:
                self.h = self.w_save
        self.d = 3
                
    def right(self):
        if self.x < self.maxx - self.w:
            self.x += self.v
            if self.w != self.h_save:
                self.w = self.h_save
            if self.h != self.w_save:
                self.h = self.w_save
        self.d = 4
```

```python
"""
    méthode propre clear screen:
    pygame.draw.rect(win, (0, 0, 0), (x, y, w, h))
    pygame.draw.rect(win, (0, 0, 0), (x, y, w, h))
    if keys[pygame.K_LEFT] and x > v:
        x -= v
        if w != h_save:
            w = h_save
        if h != w_save:
            h = w_save
            
    if keys[pygame.K_RIGHT] and x < screen_width - w:
        x += v
        if w != h_save:
            w = h_save
        if h != w_save:
            h = w_save
        
    if keys[pygame.K_UP] and y > v:
        y -= v
        if w != w_save:
            w = w_save
        if h != h_save:
            h = h_save
        
    if keys[pygame.K_DOWN] and y < screen_height - h:
        y += v
        if w != w_save:
            w = w_save
        if h != h_save:
            h = h_save
    """  
```

```python
run = True
e = Elipse(0, screen_width, 0, screen_height, 10, 15)
i = 0
s = ""
while run:
    pygame.time.delay(1000)
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
    
    keys = pygame.key.get_pressed() #https://www.pygame.org/docs/ref/key.html
    
    pygame.draw.ellipse(win, (0, 0, 0), e.getCoords())
    print(e.getCenter(), e.getHead())
    pygame.draw.line(win, (0, 0, 0), e.getCenter(), e.getHead())
    d = random.randint(1,4)
    if d == 1:
        e.up()
    if d == 2:
        e.down()
    if d == 3:
        e.left()
    if d == 4:
        e.right()
    pygame.draw.ellipse(win, e.getColor(), e.getCoords())
    pygame.draw.line(win, (255,255,255), e.getCenter(), e.getHead())
  
    if keys[pygame.K_q]:
        run = False
        
    #win.fill((0,0,0)) #clear screen
    #pygame.draw.rect(win, (255, 0, 0), (x, y, w, h)) 
    #https://www.pygame.org/docs/ref/draw.html
    
    label = myfont.render(s, 1, (0,0,0))
    win.blit(label, (0, 0))
    i += 1
    s = "Tick :"+str(i)
    label = myfont.render(s, 1, (0,255,0))
    win.blit(label, (0, 0))
    
    pygame.display.update()
pygame.quit()
```
